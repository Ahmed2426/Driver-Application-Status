```python
plt.figure(figsize=(12, 6))
plt.hist(car['App Date'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Application Dates')
plt.xlabel('Application Date')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
car['FRU Interview Scheduled'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of FRU Interview Scheduled')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(car['Drug Test'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Drug Test Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()

plt.show()
```
